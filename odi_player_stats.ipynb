```python
six = df['6s'].sum()
four = df['4s'].sum()
data=[six,four]
boundary=['6s','4s']
```

```python
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)

# to make a seperate pie
explode = (0.1,0)

plt.pie(data,autopct = lambda pct: func(pct, data),labels=boundary,explode=explode)
plt.legend(boundary,title ="no. of boundary %",loc ="lower left",bbox_to_anchor =( 0.5, 1))
# plt.title("no. of boundary percentage")
plt.show()
```

top 10 most scoring bater

```python
df_first10 = df.head(10)
plt.figure(figsize=(20,6))
plt.bar(df_first10['Player'],df_first10['Runs'])
plt.title('total Runs scorrer', fontsize=14)
plt.xticks(rotation=45)
plt.xlabel('names of cricketer')
plt.ylabel('Runs')

for i, v in enumerate(df_first10['Runs']):
    plt.text(i, v + 2, str(v), color='black', fontweight='bold',fontsize=15)
```

```python
df.sort_values(by=['Ave', 'Player'], ascending=[False, True], inplace=True)
df_top10_average = df.head(12)
df_top10_average

plt.figure(figsize=(15,6))
plt.bar(df_top10_average['Player'],df_top10_average['Ave'])
plt.xticks(rotation=45)
plt.xlabel('names of cricketer')
plt.ylabel('Average')
plt.title("top 10 best batting average")

for i, v in enumerate(df_top10_average['Ave']):
    plt.text(i, v + 2, str(v), color='black', fontweight='bold',fontsize=13)
```

```python
df.sort_values(by=['HS','Player'], ascending=[False, True], inplace=True)
df_hs = df.head(10)
df_hs

plt.figure(figsize=(15,6))
plt.bar(df_hs['Player'],df_hs['HS'])
plt.xticks(rotation=45)
plt.xlabel('names of cricketer')
plt.ylabel('Highest score')
plt.title("top 10 highest batting score(has more than 5000 runs in ODI)")

for i, v in enumerate(df_hs['HS']):
    plt.text(i, v + 2, str(v), color='black', fontweight='bold',fontsize=13)
```
